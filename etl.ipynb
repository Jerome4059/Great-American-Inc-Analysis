```python
# Dependencies
import pandas as pd
import glob
import os
from datetime import datetime
```

```python
#Read Vendor excel file into pandas
file = 'Resources/Vendor code to name.xlsx'
vendor_df = pd.read_excel(file)
# Read and Merge Monthly sales data
files = os.path.join('Resources/Data*.csv')
files = glob.glob(files)
df = pd.concat(map(pd.read_csv, files))
print(df)
```

```python
df.head()
```

```python
#Drop the first 2 Header rows
df = df.rename(columns=df.iloc[1]).drop(df.index[1]).reset_index(drop=True)
```

```python
df= df.drop(df.index[0]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.shape
```

```python
# Identify incomplete rows
df.count()
```

```python
# Drop all rows with missing information
clean_df = df.dropna(how='any').reset_index(drop=True)
```

```python
# Verify dropped rows
clean_df.count()
```

```python
clean_df.head(10)
```

```python
# Remove any features we won't be using in our analysis
clean_df.drop(['Part Number With Vendor', 'Options'], axis=1, inplace=True)
clean_df.head()
```

```python
#checking thhe counterman column for any abnormal values
clean_df['Counterman'].unique()
```

```python
#remove the anomaly
clean_df = clean_df[clean_df['Counterman'] != "!WEB"]
```

```python
#verify drop
clean_df['Counterman'].unique()
```

```python
clean_df.shape
```

```python
#Since you technically can't order a negative item i removed any row with a - in the Qty col
clean_df = clean_df[clean_df['Qty Ship'].str[0] != "-"]
```

```python
#verify drop
clean_df.shape
```

```python
clean_df.head(20)
```

```python
clean_df['Qty Ship'].unique()
```

```python
#remove money sign and replace all numbers with a parentheses with a negative sign
clean_df['Extended Part Profit'] = clean_df['Extended Part Profit'
].replace( '[\$,)]','', regex=True ).replace( '[(]','-',   regex=True ).astype(float)
```

```python
clean_df['Unit Cost'] = clean_df['Unit Cost'
].replace( '[\$,)]','', regex=True )
```

```python
clean_df['Unit Sell Price'] = clean_df['Unit Sell Price'
].replace( '[\$,)]','', regex=True )
```

```python
clean_df['Extended Part Price'] = clean_df['Extended Part Price'
].replace( '[\$,)]','', regex=True )
```

```python
clean_df.head(72)
```

```python
#Clean up Time format
clean_df['ENTRY TIME']= clean_df['ENTRY TIME'].str.replace(r'(AM|PM)', r' \1')
```

```python
clean_df.head()
```

```python
#Convert 12 hour format to 24 
clean_df['ENTRY TIME'] = pd.to_datetime(clean_df['ENTRY TIME'], errors='coerce', format= '%I:%M:%S %p').dt.strftime('%H:%M:%S')
print(clean_df)
```

```python
#veryify format conversion
clean_df.head(14)
```

```python
clean_df['ENTRY TIME'] = pd.to_timedelta(clean_df['ENTRY TIME'])
```

```python
#Bin and categorize our time column 1am-12pm(M),12pm-5pm(A),5pm-9pm(E), 9pm-12am(N)
b = pd.to_timedelta(['00:00:00','12:00:00','17:00:00','21:00:00', '24:00:00'])
l = ['Morning','Afternoon','Evening','Night']
clean_df['time of day'] = pd.cut(clean_df['ENTRY TIME'], bins=b, labels=l)
```

```python
clean_df['ENTRY TIME'] = clean_df['ENTRY TIME'].astype(str).map(lambda x: x[7:])
```

```python
clean_df.head(16)
```

```python
#reorder columns in df
clean_df = clean_df[['Ship To', 'Counterman', 'Invoice Number',
 'INVOICE DATE', 'ENTRY TIME','time of day','Part Number','Unit Cost','Qty Ship','Unit Sell Price',
 'Extended Part Price','Extended Part Profit','VND']]
```

```python
clean_df
```

```python
#Clean up Vendor name dataset
vendor_df.head(20)
```

```python
#identify incomplete rows 
vendor_df.count()
```

```python
# Drop all rows with missing information
new_vendor_df = vendor_df.dropna(how='any').reset_index(drop=True)
```

```python
# verify dropped rows
new_vendor_df.count()
```

```python
# rename column for merging purposes
new_vendor_df = new_vendor_df.rename(columns = {'Ven': 'VND'})
new_vendor_df.head()
```

```python
#merge Vendor column to our monthly sales dataset
merge_df = pd.merge(clean_df, new_vendor_df, on="VND", how="left")
merge_df.head()
```

```python
merge_df.shape
```

```python
#Drop all rows with missing Vendor name information
merge_df = merge_df.dropna(how='any').reset_index(drop=True)
```

```python
#verify drop
merge_df.shape
```

```python
#Export Dataframe as an excel file for Analysis
merge_df.to_excel("GA.xlsx", index=False)
```
